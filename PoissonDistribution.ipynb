####Распределение Пуассона

Распределение Пуассона - распределение дискретного типа случайной величины, представляющей собой число событий, произошедших за фиксированное время, при условии, что данные события происходят с некоторой фиксированной средней интенсивностью и независимо друг от друга.

Вероятность того, что ξ=k описывается формулой:
$$
\begin{align}
P(ξ=k)  = \frac{λ^ke^{-k}}{k!}
\end{align}
$$

Здесь лямбда можно понимать как n*p, что то же самое, что и Eξ. Число е выбрано по всей видимости, чтобы сумма вероятностей была равна 1

```python
fig, ax = plt.subplots(5,2,figsize=(12, 18))
plt.rc('font', size=7)

n = 100
kArray = np.arange(n)
probability = [0.1, 0.3, 0.5, 0.7, 0.9]

for i in range(0,5):
    l = n*probability[i]
    pArray = np.array([l**x * math.e**(-l) / math.factorial(x) for x in kArray])
    determArray = np.array([sum(pArray[0:x]) for x in range(n)])
    ax[i,0].scatter(kArray, pArray, marker='.', color='green')
    ax[i,0].set_title(f'Probability mass function where lambda = {int(l)}')
    ax[i,1].scatter(kArray, determArray, marker='.', color='blue')
    ax[i,1].set_title(f'Cumulative distribution function where lambda = {int(l)}')
```
